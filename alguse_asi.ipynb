```python
import pandas as pd
import numpy as np
import datetime as dt # for date data type
import matplotlib.pyplot as plt
import numpy as np
import pytz
```

```python
# Read the Johvi weather data XLSX file into a DataFrame
johvi = pd.read_excel('https://www.ilmateenistus.ee/wp-content/uploads/2023/02/Johvi.xlsx', header=2)
# Make new united column of datetime
johvi['datetime_str'] = johvi['Aasta'].astype(str) + '-' + johvi['Kuu'].astype(str) + '-' + johvi['Päev'].astype(str) + ' ' + johvi['Kell (UTC)'].astype(str)
johvi['datetime'] = pd.to_datetime(johvi['datetime_str'], format='%Y-%m-%d %H:%M:%S')
time_zone = pytz.timezone('Europe/Tallinn')
johvi['datetime'] = johvi['datetime'].dt.tz_localize('UTC').dt.tz_convert(time_zone)
johvi.drop('datetime_str', axis=1, inplace=True)

# Read the Tartu weather data XLSX file into a DataFrame
tartu = pd.read_excel('https://www.ilmateenistus.ee/wp-content/uploads/2023/02/Tartu-Toravere.xlsx', header=2)
# Make new united column of datetime
tartu['datetime_str'] = tartu['Aasta'].astype(str) + '-' + tartu['Kuu'].astype(str) + '-' + tartu['Päev'].astype(str) + ' ' + tartu['Kell (UTC)'].astype(str)
tartu['datetime'] = pd.to_datetime(tartu['datetime_str'], format='%Y-%m-%d %H:%M:%S')
time_zone = pytz.timezone('Europe/Tallinn')
tartu['datetime'] = tartu['datetime'].dt.tz_localize('UTC').dt.tz_convert(time_zone)
tartu.drop('datetime_str', axis=1, inplace=True)

# Read the Tallinn weather data XLSX file into a DataFrame
tallinn = pd.read_excel('https://www.ilmateenistus.ee/wp-content/uploads/2023/02/Tallinn-Harku.xlsx', header=2)
# Make new united column of datetime
tallinn['datetime_str'] = tallinn['Aasta'].astype(str) + '-' + tallinn['Kuu'].astype(str) + '-' + tallinn['Päev'].astype(str) + ' ' + tallinn['Kell (UTC)'].astype(str)
tallinn['datetime'] = pd.to_datetime(tallinn['datetime_str'], format='%Y-%m-%d %H:%M:%S')
time_zone = pytz.timezone('Europe/Tallinn')
tallinn['datetime'] = tallinn['datetime'].dt.tz_localize('UTC').dt.tz_convert(time_zone)
tallinn.drop('datetime_str', axis=1, inplace=True)

# Read the Virtsu weather data XLSX file into a DataFrame
virtsu = pd.read_excel('https://www.ilmateenistus.ee/wp-content/uploads/2023/02/Virtsu.xlsx', header=2)
# Make new united column of datetime
virtsu['datetime_str'] = virtsu['Aasta'].astype(str) + '-' + virtsu['Kuu'].astype(str) + '-' + virtsu['Päev'].astype(str) + ' ' + virtsu['Kell (UTC)'].astype(str)
virtsu['datetime'] = pd.to_datetime(virtsu['datetime_str'], format='%Y-%m-%d %H:%M:%S')
time_zone = pytz.timezone('Europe/Tallinn')
virtsu['datetime'] = virtsu['datetime'].dt.tz_localize('UTC').dt.tz_convert(time_zone)
virtsu.drop('datetime_str', axis=1, inplace=True)
```

```python
# drop unimportant columns and rename temperature column
johvi = johvi.rename(columns={'Õhutemperatuur °C': 'temp_johvi'})
johvi = johvi[['datetime', 'temp_johvi']]
tartu = tartu.rename(columns={'Õhutemperatuur °C': 'temp_tartu'})
tartu = tartu[['datetime', 'temp_tartu']]
tallinn = tallinn.rename(columns={'Õhutemperatuur °C': 'temp_tallinn'})
tallinn = tallinn[['datetime', 'temp_tallinn']]
virtsu = virtsu.rename(columns={'Õhutemperatuur °C': 'temp_virtsu'})
virtsu = virtsu[['datetime', 'temp_virtsu']]
```

```python
# Merge all weather dataframes to one
merged_df = pd.merge(johvi, tartu, on='datetime', how='inner')
merged_df = pd.merge(merged_df, tallinn, on='datetime', how='inner')
merged_df = pd.merge(merged_df, virtsu, on='datetime', how='inner')
merged_df
```

```python
from shapely.geometry import Point
import pyproj


# Convert data types to date/time, int etc
veerud_int=['Isikuid', 'Hukkunuid', "Vigastatuid", "Sõidukeid", "Kergliikurijuhi osalusel", "Jalakäija osalusel", "Kaassõitja osalusel", "Maastikusõiduki juhi osalusel", "Eaka (65+) mootorsõidukijuhi osalusel", "Bussijuhi osalusel", "Veoautojuhi osalusel", "Ühissõidukijuhi osalusel", "Sõiduautojuhi osalusel", "Mootorratturi osalusel", "Mopeedijuhi osalusel", "Jalgratturi osalusel", "Alaealise osalusel", "Turvavarustust mitte kasutanud isiku osalusel", "Esmase juhiloa omaniku osalusel", "Mootorsõidukijuhi osalusel", "Lubatud sõidukiirus (PPA)", "Tee nr (PPA)",  "GPS X", "GPS Y"]
veerud_cat=["Maakond (PPA)", "Omavalitsus (PPA)", "Asustus (PPA)", "Liiklusõnnetuse liik [1]", "Liiklusõnnetuse liik [3]", "Tüüpskeemi nr",  "Tüüpskeem [2]", "Tee tüüp [1]", "Tee tüüp [2]", "Tee element [1]", "Tee element [2]", "Tee objekt [2]", "Kurvilisus", "Tee tasasus", "Tee seisund", "Teekate", "Teekatte seisund [2]", "Sõiduradade arv", "Ilmastik [1]", "Valgustus [1]", "Valgustus [2]"]

df["datetime"]=pd.to_datetime(df["Toimumisaeg"], format='mixed')
df[veerud_int] = df[veerud_int].astype('Int64', errors='ignore')
df[veerud_cat]=df[veerud_cat].astype("category", errors='ignore')



# Drop rows without GPS coordinates
df.dropna(subset=['GPS Y', 'GPS Y'], inplace=True)

# Drop rows without Time
df.dropna(subset=['datetime'], inplace=True)

# Drop rows with unrealistic speed limit
df = df[(df["Lubatud sõidukiirus (PPA)"] >= 0) & (df["Lubatud sõidukiirus (PPA)"] <= 120) & (df["Lubatud sõidukiirus (PPA)"].notna())]

df['Asukoht'] = df.apply(lambda row: Point(float(row['GPS Y']), float(row['GPS X']))
                          if not pd.isna(row['GPS X']) and not pd.isna(row['GPS Y']) else None,
                          axis=1)

# Define the source and target CRS
source_crs = pyproj.CRS.from_epsg(3301)  # EPSG 3301
target_crs = pyproj.CRS.from_epsg(4326)  # EPSG 4326

# Create a transformer to convert coordinates from source CRS to target CRS
transformer = pyproj.Transformer.from_crs(source_crs, target_crs, always_xy=True)

# Transform the Point coordinates
df['Asukoht'] = df['Asukoht'].apply(lambda point: Point(transformer.transform(point.x, point.y)) if point is not None else None)

# outside Estonia - possible outliers
# Define a bounding box for Estonia's latitude and longitude
estonia_bbox = {
    "lat_min": 57.509868,
    "lat_max": 59.675660,
    "lon_min": 21.382507,
    "lon_max": 28.209960
}

# Filter rows where Asukoht is within Estonia's bounding box
df = df[
    (df["Asukoht"].apply(lambda point: point.x) >= estonia_bbox["lon_min"]) &
    (df["Asukoht"].apply(lambda point: point.x) <= estonia_bbox["lon_max"]) &
    (df["Asukoht"].apply(lambda point: point.y) >= estonia_bbox["lat_min"]) &
    (df["Asukoht"].apply(lambda point: point.y) <= estonia_bbox["lat_max"])
]

df

```

```python
import re
from math import radians, sin, cos, sqrt, atan2

# Create a DataFrame with the weather measurment points
targets = pd.DataFrame({
    'Location': ['temp_johvi', 'temp_virtsu', 'temp_tallinn', 'temp_tartu'],
    'Latitude': [59.328889, 58.574722, 59.398056, 58.264167],
    'Longitude': [27.398333, 23.51, 24.602778, 26.461389]
})

# Create a helper function to calculate the distance between two points using the Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    a = sin(dlat/2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

# Iterate over each row in the DataFrame and calculate the distances to each target point
for index, row in df.iterrows():
    min_distance = float('inf')
    nearest_location = ''
    latitude, longitude = row['Asukoht'].y, row['Asukoht'].x
    if latitude is not None and longitude is not None:
        for _, target_row in targets.iterrows():
            distance = calculate_distance(latitude, longitude, target_row['Latitude'], target_row['Longitude'])
            if distance < min_distance:
                min_distance = distance
                nearest_location = target_row['Location']
    # Add the nearest location to a new column in the DataFrame
    df.loc[index, 'Nearest_Location'] = nearest_location
```

```python
# Create a new column "Temperature" in your original DataFrame
df['Temperature'] = None


# Iterate over each row in the DataFrame and find temperature value from merged_df
for index, row in df.iterrows():
    datetime_value = row['datetime']
    timezone = pytz.timezone('Europe/Tallinn')
    datetime_value = timezone.localize(datetime_value)
    nearest_location = row['Nearest_Location']
    nearest_row = merged_df.loc[(merged_df['datetime'] - datetime_value).abs().idxmin()]
    temperature_value = nearest_row[nearest_location]
    df.at[index, 'Temperature'] = temperature_value
```
